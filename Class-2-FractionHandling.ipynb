```python

class Fraction: 

    def __init__(self,n,d): 
        self.numi = n 
        self.deno = d 
    def __str__(self):    # this is a magic method when we create a object it runs automatically(return as sting)
        return f"{self.numi}/{self.deno}"
    def gcd(self,a,b): 
        while b != 0:
          a,b = b,a % b
        return a

    def __add__(self,other):  # this one also a magic method for addition
        new_numi = self.numi*other.deno + other.numi*self.deno 
        new_deno = self.deno*other.deno
        temp = self.gcd(new_numi,new_deno)
        return f"{new_numi//temp}/{new_deno//temp}"

    def __sub__(self,other):      # magic method substraction
        new_numi = self.numi*other.deno - other.numi*self.deno 
        new_deno = self.deno*other.deno

        temp = self.gcd(new_numi,new_deno)
        return f"{new_numi//temp}/{new_deno//temp}"

    def __mul__(self,other):     # magic method for multiplication
        new_numi = self.numi*other.numi
        new_deno = self.deno*other.deno

        temp = self.gcd(new_numi,new_deno)
        return f"{new_numi//temp}/{new_deno//temp}"

    def __truediv__(self,other): 
        new_numi = self.numi*other.deno
        new_deno = self.deno*other.numi

        temp = self.gcd(new_numi,new_deno)
        return f"{new_numi//temp}/{new_deno//temp}"
```

```python

f1=Fraction(2,3)
f2=Fraction(4,3)
add=f1+f2
sub=f1-f2
mul=f1*f2 
div=f1/f2

print(add,",",sub,",",mul,",",div)
```
